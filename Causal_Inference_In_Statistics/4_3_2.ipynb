a) We calculate the regression $Y = r_{1}X + r_{2}H$. This will us the direct effects of $X$ and $H$ on $Y$, specifically $r_{1} = b$ and $r_{2} = c$.

Next we calculate the regression $Y = r_{3}X$ which will give us the total effect of $X$ on $Y$. According to the model, $r_{3} = ac + b$. Therefore, we can solve for $a$

$$
a = \frac{r_{3} - r_{1}}{r_{2}}
$$

b)
$$\begin{align}
E[Y_{1} - Y_{0} \mid Y = 1] &= E[Y_{1}\mid Y=1] - E[Y_{0}\mid Y=1]\\
&= E[Y\mid Y=1] + \tau(1 - E[X\mid Y=1]) - E[Y\mid Y=1] - \tau(0 - E[X\mid Y=1])\\
&= 1 + \tau - \tau E[X\mid Y=1] - 1 + \tau E[X\mid Y=1]\\
&= \tau\\
&= ab
\end{align}$$

c)
We have the following structural model

$$\begin{align}
&H = U_{H}\\
&X = aH + U_{X}\\
&Y = bX + cH + \delta XH + U_{Y}
\end{align}$$

First we calculate $\tau = E[Y\mid \textit{do}(X=x+1)] - E[Y\mid \textit{do}(X=x)]$.

$$\begin{align}
E[Y\mid \textit{do}(X=x+1)] &= E[b(x+1) + cH + \delta (x+1)H + U_{Y}]\\
&= E[bx + b + cU_{H} + \delta xU_{H} + \delta U_{H} + U_{Y}]\\
&= bx + b + cE[U_{H}] + \delta x E[U_{H}] + \delta E[U_{H}] + E[U_{Y}]\\
&= bx + b
\end{align}$$

$$\begin{align}
E[Y\mid \textit{do}(X=x)] &= E[bx + cH + \delta xH + U_{Y}]\\
&= E[bx + cU_{H} + \delta xU_{H} + U_{Y}]\\
&= bx + cE[U_{H}] + \delta x E[U_{H}] + E[U_{Y}]\\
&= bx
\end{align}$$

$$\implies \tau = bx + b - bx = b$$

Next we calculate $\textit{ETT} = E[Y_{1} - Y_{0} \mid X=1]$. Using the basic definition of counterfactuals and the equality $E[H \mid X = 1] = R_{HX}$

$$\begin{align}
E[Y_{1} \mid X=1] &= E[b + cR_{HX} + \delta R_{HX} + U_{Y}]\\
&= b + cR_{HX} + \delta R_{HX}
\end{align}$$

$$\begin{align}
E[Y_{0} \mid X=1] &= E[cR_{HX} + U_{Y}]\\
&= cR_{HX}
\end{align}$$

$$\implies \textit{ETT} =b + cR_{HX} + \delta R_{HX} - cR_{HX} = b + \delta R_{HX}$$

$$\begin{align}
\sigma_{HX} &= E[(H - E[H])(X - E[X])]\\
&= E[HX]\\
&= E[aU^2_{H} + U_{X}U_{H}]\\
&= a
\end{align}$$

$$\begin{align}
\sigma^2_{X} &= E[X^2] - E[X]^2\\
&= E[X^2]\\
&= E[(aU_{H} + U_{X})(aU_{H} + U_{X})]\\
&= E[a^2U^2_{H} + 2aU_{X}U_{H} + U^2_{X}]\\
&= a^2 + 1
\end{align}$$

$$\implies R_{HX} = \frac{\sigma_{HX}}{\sigma^2_{X}} = \frac{a}{a^2 + 1}$$

In the above we assume that $U_{H}$ and $U_{X}$ are both $N(0,1)$ and therefore $E[U^2_{H}] = E[U^2_{X}] = 1$.

This gives us

$$\textit{ETT} = b + \frac{\delta a}{a^2 + 1}$$

Finally, putting it all together

$$\tau - \textit{ETT} = b - b - \frac{\delta a}{a^2 + 1} = - \frac{\delta a}{a^2 + 1}$$

Which has the opposite sign to what is given in the book, so one of us must be wrong.
