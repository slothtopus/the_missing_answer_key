# Study Question 3.3.3

a) Both being assigned the treatment $T$ and being depressed $D$ influence the likelihood of receiving a lollipop $L$. Being depressed and treatment both influence recovery $R$. This gives us the causal graph below.

```python
from graphviz import Digraph

dot = Digraph()

dot.node('T')
dot.node('D')
dot.node('L')
dot.node('R')

dot.edge('T', 'L')
dot.edge('D', 'L')
dot.edge('T', 'R')
dot.edge('D', 'R')

dot
```

b) There is no backdoor path between $T$ and $R$. However, adjusting for $L$ would open up the collider $T \rightarrow L \leftarrow D$ and create a backdoor path.

c) Since the graph does not contain any backdoor paths $P(R \mid \textit{do}(T)) = P(R \mid T)$

d) The question doesn't specify whether the lollipops were also given out to depressed people as well the day after treatment. Let's assume they were and that being depressed before treatment $D_{\textit{before}}$ influences the likelihood of being depressed after treatment $D_{\textit{after}}$, but only being depressed beforehand influences recovery. This gives us the graph below.

```python
dot = Digraph()

dot.node('T')
dot.node('D_before')
dot.node('D_after')
dot.node('L')
dot.node('R')

dot.edge('T', 'L')
dot.edge('D_after', 'L')
dot.edge('T', 'R')
dot.edge('D_before', 'R')
dot.edge('D_before', 'D_after')

dot
```

Here there is again no backdoor path from $T$ to $R$ so we have nothing to adjust for. However, once again, adjusting for $L$ would open up the backdoor path $T \rightarrow L \leftarrow D_{\textit{after}} \leftarrow D_{\textit{before}} \rightarrow R$, so should be avoided.

Since there's no backdoor path, again we have  $P(R \mid \textit{do}(T)) = P(R \mid T)$.
