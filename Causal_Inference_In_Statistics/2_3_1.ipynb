f) We can rephrase this as series of easier questions:

- Are $Y$ and $X$ independent given $R,S,T,P$ - yes
- Are $Y$ and $R$ independent given $X,S,T,P$ - yes
- Are $Y$ and $S$ independent given $X,R,T,P$ - no
- Are $Y$ and $P$ independent given $X,R,S,T$ - yes

Independence conditional on the other variables implies a zero coefficient. Therefore, only $S$ would have a non-zero coefficient.
