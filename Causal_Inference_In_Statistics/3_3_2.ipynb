a) From plot of the data in Figure 3.9, it is clear that the choice of plan is dependent on the student's initial weight. We can express this in our graph as an unmeasured node $S$, which represents factors about the student that effect both the their initial weight and their choice of plan.

To make the question easier, we can also assume that the final weight $W_{f}$ is dependent only on the initial weight $W_{i}$ and not on both $S$ and $W_{i}$.

Using $P$ to represent the choice of plan, this gives us the following graph.

```python
from graphviz import Digraph

dot = Digraph()

dot.node('S')
dot.node('P')
dot.node('Wi')
dot.node('Wf')
dot.node('Wf - Wi')

dot.edge('S', 'P')
dot.edge('S', 'Wi')
#dot.edge('S', 'Wf')
dot.edge('P', 'Wf')
dot.edge('Wi', 'Wf')
dot.edge('Wi', 'Wf - Wi')
dot.edge('Wf', 'Wf - Wi')

dot
```

b) The first statistician calculated $P(W_{f} - W_{i} \mid P)$. There is a backdoor path from $P$ to $W_{f} - W_{i}$, namely $P \rightarrow S \rightarrow W_{i} \rightarrow W_{f} - W_{i}$, so simply conditioning on $P$ will give the wrong result.

This path can be blocked by conditioning on $W_{i}$, which would delete the arrow going from $S \rightarrow W_{i}$. This is what the second statistician did when segmenting the students by their initial weight. Therefore, the second statistician was right!

c) If you shift the students in bubble $A$ so that the average distance to the $W_{f} = W_{i}$ line was positive and the students in bubble $B$ so that the average distance was negative, just looking at $W_{f} - W_{i}$ would show that students in $A$ gained weight, but students in $B$ lost weight. Provided this shift was small enough, the average value of $B$ could still be greater than $A$ for every constant starting weight $W_{0}$, meaning that students in $B$ gained more weight than in $A$. Thus we have an example of Simpson's reversal when we condition on starting weight $W_{i}$.
