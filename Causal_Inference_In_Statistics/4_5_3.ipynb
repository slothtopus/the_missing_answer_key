# Study Question 4.5.3

a) Natural Direct Effect:

$$\begin{align}
\textit{NDE} &= E[Y_{1,M_{0}} - Y_{0,M_{0}}]\\
&=E[Y_{1,M_{0}}] - E[Y_{0,M_{0}}]
\end{align}$$

$$\begin{align}
Y_{1,M_{0}} \implies &t = 1\\
&w = \alpha + u_{w}\\
&m = \gamma_{2}u_{w} + u_{m}\\
&y = \beta_{1}\gamma_{2}u_{w} + \beta_{1}u_{m} + \beta_{2} + \beta_{3}\gamma_{2}u_{w} + \beta_{3}u_{m} + \beta_{4}\alpha + \beta_{4}u_{w} + u_{y}\\
\implies &E[y] = \beta_{2} + \beta_{4}\alpha
\\
\\
Y_{0,M_{0}} \implies &t = 0\\
&w = u_{w}\\
&m = \gamma_{2}u_{w} + u_{m}\\
&y = \beta_{1}\gamma_{2}u_{w} + \beta_{1}u_{m} + \beta_{4}u_{w} + u_{y}\\
\implies &E[y] = 0
\end{align}$$

$$\implies \textit{NDE} = \beta_{2} + \beta_{4}\alpha$$

Natural Indirect Effect:

$$\begin{align}
\textit{NIE} &= E[Y_{0,M_{1}} - Y_{0,M_{0}}]\\
&=E[Y_{0,M_{1}}] - E[Y_{0,M_{0}}]
\end{align}$$

$$\begin{align}
Y_{0,M_{1}} \implies &t = 0\\
&w = u_{w}\\
&m = \gamma_{1} + \gamma_{2}\alpha + \gamma_{2}u_{w} + u_{m}\\
&y = \beta_{1}\gamma{1} + \beta_{1}\gamma_{2}\alpha + \beta_{1}\gamma_{2}u_{w} + \beta_{1}u_{m} + \beta_{4}u_{w} + u_{y}\\
\implies &E[y] = \beta_{1}\gamma_{1} + \beta_{1}\gamma_{2}\alpha
\end{align}$$

$$\implies \textit{NIE} = \beta_{1}(\gamma_{1} + \gamma_{2}\alpha)$$

Total Effect:

$$\begin{align}
\textit{TE} &= E[Y_{1} - Y_{0}]\\
&=E[Y_{1}] - E[Y_{0}]
\end{align}$$

$$\begin{align}
Y_{1} \implies &t = 1\\
&w = \alpha + u_{w}\\
&m = \gamma_{1} + \gamma_{2}\alpha + \gamma_{2}u_{w} + u_{m}\\
&y = \beta_{1}\gamma_{1} + \beta_{1}\gamma_{2}\alpha + \beta_{1}\gamma_{2}u_{w} + \beta_{1}u_{m} + \beta_{2} + \beta_{3}\gamma_{1} + \beta_{3}\gamma_{2}\alpha + \beta_{3}\gamma_{2}u_{w} + \beta_{3}u_{m} + \beta_{4}\alpha + \beta_{4}u_{w} + u_{y}\\
\implies &E[y] = \beta_{1}\gamma_{1} + \beta_{1}\gamma_{2}\alpha + \beta_{2} + \beta_{3}\gamma_{1} + \beta_{3}\gamma_{2}\alpha + \beta_{4}\alpha\\
\\
\\
Y_{0} \implies &t = 0\\
&w = u_{w}\\
&m = \gamma_{2}u_{w} + u_{m}\\
&y = \beta_{1}\gamma_{2}u_{w} + \beta_{1}u_{m} + \beta_{4}u_{w} + u_{y}\\
\implies &E[y] = 0
\end{align}$$

$$\begin{align}
\implies \textit{TE} &= \beta_{1}\gamma_{1} + \beta_{1}\gamma_{2}\alpha + \beta_{2} + \beta_{3}\gamma_{1} + \beta_{3}\gamma_{2}\alpha + \beta_{4}\alpha\\
&= (\beta_{1} + \beta_{3})(\gamma_{1} + \gamma_{2}\alpha) + \beta_{2} + \beta_{4}\alpha
\end{align}$$

$$\implies \textit{TE} - \textit{NDE} = (\beta_{1} + \beta_{3})(\gamma_{1} + \gamma_{2}\alpha)$$
