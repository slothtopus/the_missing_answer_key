# Study Question 4.5.2

a) We have the following structural model

$$\begin{align}
y &= \beta_{1}m + \beta_{2}t + U_{y}\\
m &= \gamma_{1}t + U_{m}
\end{align}$$

Natural Direct Effect:

$$\begin{align}
\textit{NDE} &= E[Y_{1,M_{0}} - Y_{0,M_{0}}]\\
&=E[Y_{1,M_{0}}] - E[Y_{0,M_{0}}]
\end{align}$$

$$\begin{align}
E[Y_{1,M_{0}}] &\implies t = 1, m = U_m\\
&\implies y = \beta_{1}U_{m} + \beta_{2} + U_{y}\\
&\implies E[y] = 0 + \beta_{2} + 0 = \beta_{2}\\
\end{align}$$

$$\begin{align}
E[Y_{0,M_{0}}] &\implies t = 0, m = U_m\\
&\implies y = \beta_{1}U_{m} + U_{y}\\
&\implies E[y] = 0 + 0 = 0\\
\end{align}$$

$$\begin{align}
\implies \textit{NDE} &= \beta_{2} - 0\\
&= \beta_{2}
\end{align}$$

Natural Indirect Effect:

$$\begin{align}
\textit{NIE} &= E[Y_{0,M_{1}} - Y_{0,M_{0}}]\\
&=E[Y_{0,M_{1}}] - E[Y_{0,M_{0}}]
\end{align}$$

$$\begin{align}
E[Y_{0,M_{1}}] &\implies t = 0, m = \gamma_{1} + U_{m}\\
&\implies y = \beta_{1}\gamma_{1} + \beta_{1}U_{m} + U_{y}\\
&\implies E[y] = \beta_{1}\gamma_{1} + 0 + 0 = \beta_{1}\gamma_{1}\\
\end{align}$$

$$\begin{align}
\implies \textit{NIE} &= \beta_{1}\gamma_{1} - 0\\
&= \beta_{1}\gamma_{1}
\end{align}$$

Total Effect

$$\begin{align}
\textit{TE} &= E[Y_{1} - Y_{0}]\\
&=E[Y_{1}] - E[Y_{0}]
\end{align}$$

$$\begin{align}
E[Y_{1}] &\implies t = 1\\
&\implies y = \beta_{1}\gamma_{1} + \beta_{1}U_{m} + \beta_{2} + U_{y}\\
&\implies E[y] = \beta_{1}\gamma_{1} + 0 + \beta_{2} + 0 = \beta_{1}\gamma_{1} + \beta_{2}\\
\end{align}$$

$$\begin{align}
E[Y_{0}] &\implies t = 0\\
&\implies y = \beta_{1}U_{m} + U_{y}\\
&\implies E[y] = 0 + 0 = 0\\
\end{align}$$

$$\begin{align}
\implies \textit{TE} &= \beta_{1}\gamma_{1} + \beta_{2} - 0\\
&= \beta_{1}\gamma_{1} + \beta_{2}
\end{align}$$

And $\textit{TE} = \textit{NDE} + \textit{NIE}$ as expected.

b) The answer remains the same if $U_{m}$ is correlated with $U_{y}$, since we did not need to take any expectations of terms involving both $U_{m}$ and $U_{y}$ in the calculations above.
