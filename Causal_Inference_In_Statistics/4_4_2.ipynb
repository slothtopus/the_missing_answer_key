a) Since Joe is about to smoke, we can put him in the category of the treated ($X=1$). Therefore, we can use $\textit{ETT}$ to estimate the increase in his risk of cancer if he smokes

$$\textit{ETT} = E[Y_{1} - Y_{0}\mid X=1]$$

b) To estimate $\textit{ETT}$ we either need a set of measured variables $Z$ that will block all backdoor paths from $X$ to $Y$. Or if $Z$ is not available we can estimate $\textit{ETT}$ if we know $P(X,Y)$ and $P(Y\mid \textit{do}(X))$

c) We do not have a suitable set of backdoor variables Z, but we can use the front-door adjustment to estimate $P(Y\mid \textit{do}(X))$

- $X = 1$ and $X = 0$ smoking and not smoking respectively
- $Z = 1$ and $Z = 0$ tar and no tar respectively
- $Y = 1$ and $Y = 0$ cancer and no cancer respectively

|X|Z|Y|n|
|:-:|:-:|:-:|:-:|
0|0|0|38
1|0|0|18
0|1|0|1
1|1|0|323

From the table we have:

$$\begin{align}
&P(X=0) = \frac{400}{800}\\
&P(X=1) = \frac{400}{800}\\
&P(Y=1\mid X=1) = \frac{59}{400}\\
&P(Y=1\mid X=0) = \frac{361}{400}\\
&P(Z=0\mid X=0) = \frac{380}{400}\\
&P(Z=1\mid X=0) = \frac{20}{400}\\
&P(Y=1\mid Z=0,X=0) = \frac{342}{380}\\
&P(Y=1\mid Z=0,X=1) = \frac{2}{20}\\
&P(Y=1\mid Z=1,X=0) = \frac{19}{20}\\
&P(Y=1\mid Z=1,X=1) = \frac{57}{380}\\
\end{align}$$

We calculate $P(Y=1\mid\textit{do}(X=0))$ using the front-door adjustment

$$\begin{align}
P(Y=1\mid\textit{do}(X=0)) &= \sum_{z}\sum_{x'}P(Y=1\mid Z=z,X=x')P(X=x')P(Z=z\mid X=0)\\
&= P(Y=1\mid Z=0,X=0)P(X=0)P(Z=0\mid X=0)\\
&\ \ \ \ + P(Y=1\mid Z=0,X=1)P(X=1)P(Z=0\mid X=0)\\
&\ \ \ \ + P(Y=1\mid Z=1,X=0)P(X=0)P(Z=1\mid X=0)\\
&\ \ \ \ + P(Y=1\mid Z=1,X=1)P(X=1)P(Z=1\mid X=0)
\\
\\
&= \frac{342}{380}\cdot\frac{400}{800}\cdot\frac{380}{400} + \frac{2}{20}\cdot\frac{400}{800}\cdot\frac{380}{400}\\
&\ \ \ \ + \frac{19}{20}\cdot\frac{400}{800}\cdot\frac{20}{400} + \frac{57}{380}\cdot\frac{400}{800}\cdot\frac{20}{400}\\
\\
&= \frac{171}{400} + \frac{19}{400} + \frac{19}{800} + \frac{3}{800}\\
\\
&= \frac{201}{400}
\end{align}$$

And then we put it all together to calculate $\textit{ETT}$

$$\begin{align}
\textit{ETT} &= P(Y=1\mid X=1) - \frac{P(Y=1 \mid \textit{do}(X=0)) - P(Y=1\mid X=0)P(X=0)}{P(X=1)}\\
&= \frac{59}{400} - \frac{\frac{201}{400} - \frac{361}{400}\cdot\frac{400}{800}}{\frac{400}{800}}\\
&= \frac{9}{200}
\end{align}$$

Joe will have a roughly 5% increased chance of cancer if he smokes, so he probably shouldn't.
