a) Using the decomposition $E[Y_{x}] = E[Y_{x}\mid x']P(x') + E[Y_{x}\mid x]P(x)$, we have

$$\begin{align}
&E[Y_{X=0}] = E[Y_{X=0}|X=0]P(X=0) + E[Y_{X=0}|X=1]P(X=1)\\
\\
\implies &E[Y_{X=0}|X=1] = \frac{E[Y_{X=0}] - E[Y_{X=0}|X=0]P(X=0)}{P(X=1)}
\end{align}$$

Since both $X$ and $Y$ are binary, we can make the following substitutions

$$\begin{align}
&E[Y_{X=x}] = P(Y=1 \mid \textit{do}(X=x))\\
\\
&E[Y_{X=x}\mid X=x] = P(Y=1 \mid X=x)
\end{align}$$

Putting it all together to calculate $\textit{ETT}$

$$\begin{align}
\textit{ETT} &= E[Y_{X=1} \mid X=1] -  E[Y_{X=0} \mid X=1]\\
\\
&=P(Y=1\mid X=1) - \frac{P(Y=1 \mid \textit{do}(X=0)) - P(Y=1\mid X=0)P(X=0)}{P(X=1)}
\end{align}$$

The quantities above are all estimable if $P(X=x, Y=y)$ and $P(Y=y\mid \textit{do}(X=x))$ are known.

From the table

$$\begin{align}
&P(Y=1\mid X=0,Z=0) = \frac{234}{270}\\

&P(Y=1\mid X=1) = \frac{273}{350}\\

\end{align}$$

We need to estimate $P(Y=1\mid \textit{do}(X=0))$, which we can get using the adjustment formula with $Z$

$$\begin{align}
P(Y=1\mid \textit{do}(X=0)) &= \sum_{z} P(Y=1\mid X=0,Z=z)P(Z=z)\\
&= P(Y=1\mid X=0,Z=0)P(Z=0) + P(Y=1\mid X=0,Z=1)P(Z=1)\\

\end{align}$$

Using the formula from part a) we have

$$\begin{align}
\textit{ETT} &= \frac{273}{350} - \frac{\frac{6231}{8000} - \frac{289}{350}\cdot\frac{350}{700}}{\frac{350}{700}}\\
&=\frac{1343}{28000}
\end{align}$$
